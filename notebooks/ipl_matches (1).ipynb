```python
# “Does winning the toss help in winning the match?”

matches['Toss_win_match_win']=(matches['toss_winner']==matches['winner'])
matches['Toss_win_match_win'].value_counts(normalize=True)*100
result=matches['Toss_win_match_win'].value_counts(normalize=True)*100
result.plot(kind='pie',autopct='%1.1f%%',colors=['#DDBDF2','#DCF2BD'],explode=[0.1,0])
plt.title('Toss_win_match_win')
plt.savefig('IPL-Data-Analysis/outputs/graph17.png')

```

### 🔍 Insight:
- Teams that win the toss also win the match in approximately 55–60% of cases.
- This indicates that winning the toss provides a slight advantage.
- However, since a significant percentage of teams still lose after winning the toss, match outcome largely depends on team performance rather than toss alone.

```python
# Win probability based on runs
nm=deliever.merge(matches,left_on='match_id',right_on='id')

chase=nm[nm['inning']==2]

target=nm[nm['inning']==1].groupby('match_id')['total_runs'].sum().reset_index()

target.rename(columns={'total_runs': 'target'}, inplace=True)

chase=chase.merge(target,on='match_id')

chase['current_score']=chase.groupby('match_id')['total_runs'].cumsum()

chase['run_left']=chase['target']-chase['current_score']

chase['ball_left'] = 120 - (chase['over']*6 + chase['ball'])

chase = chase[(chase['run_left'] < 20) & (chase['ball_left'] < 10)]

chase['result']=(chase['batting_team']==chase['winner']).astype(int)

prob = chase.groupby(['run_left','ball_left'])['result'].mean().reset_index()
sample=prob.sample(10)
sample
```

```python
plt.plot(sample.index,sample.values)
plt.title('Win probability based on runs')
plt.xticks(rotation='vertical')
plt.savefig('IPL-Data-Analysis/outputs/graph18.png')
```

“The analysis reveals that IPL matches are highly volatile in the final overs, where even small differences in required runs can drastically change win probability, making death overs the most निर्णायक phase of the game.”

## 🏁 Final Conclusion

- IPL matches are highly dynamic, with different phases (powerplay, middle overs, death overs) playing crucial roles.
- Teams rely on aggressive batting in death overs to maximize scoring.
- Powerplay performance helps build early momentum.
- Venue conditions significantly impact scoring patterns.
